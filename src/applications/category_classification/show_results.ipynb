```python
import os
import pandas as pd

# Caminho base
base_path = "results/category_classification"

# Lista para armazenar os resultados
all_results = []

# Caminhar pelos diretórios
for root, _, files in os.walk(base_path):
    for file in files:
        if file == "best_f1_score_model.csv":
            # Caminho completo do arquivo
            file_path = os.path.join(root, file)
            
            # Extrair o jornal e o modelo do caminho
            path_parts = file_path.split("/")
            journal = path_parts[2]  # Elemento após a segunda /
            model = path_parts[3]    # Elemento após a terceira /
            
            # Ler o CSV
            df = pd.read_csv(file_path)
            
            # Selecionar as métricas de interesse
            mean_accuracy = df["mean_test_accuracy"].values[0]
            std_accuracy = df["std_test_accuracy"].values[0]
            mean_precision = df["mean_test_precision"].values[0]
            std_precision = df["std_test_precision"].values[0]
            mean_recall = df["mean_test_recall"].values[0]
            std_recall = df["std_test_recall"].values[0]
            mean_f1 = df["mean_test_f1_score"].values[0]
            std_f1 = df["std_test_f1_score"].values[0]
            
            # Combinar as métricas com seus desvios
            combined_accuracy = f"{mean_accuracy:.4f} ± {std_accuracy:.4f}"
            combined_precision = f"{mean_precision:.4f} ± {std_precision:.4f}"
            combined_recall = f"{mean_recall:.4f} ± {std_recall:.4f}"
            combined_f1 = f"{mean_f1:.4f} ± {std_f1:.4f}"
            
            # Armazenar no formato de tabela
            all_results.append({
                "Journal": journal,
                "Model": model,
                "Accuracy": combined_accuracy,
                "Precision": combined_precision,
                "Recall": combined_recall,
                "F1-Score": combined_f1
            })

# Criar um DataFrame com todos os resultados
results_df = pd.DataFrame(all_results)

# Salvar o resultado em um arquivo CSV
#results_df.to_csv("combined_results.csv", index=False)
results_df
```
